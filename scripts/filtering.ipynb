```python
import pandas as pd
import numpy as np
```

```python
df = pd.read_csv('wca_organizers_by_distance.csv')
persons = pd.read_csv('WCA_export_Persons.tsv', sep='\t')
```

```python
df = pd.merge(df, persons, left_on='wca_id', right_on='id', how='left')[['wca_id', 'name', 'gender', 'home_country', 'competition_id', 'competition_country', 'distance_km']]
```

```python
df_cleaned = df.dropna(subset=['wca_id', 'distance_km'])

df_max_distance = df_cleaned.loc[df_cleaned.groupby('wca_id')['distance_km'].idxmax()].reset_index(drop=True)
df_max_distance['distance_km'] = df_max_distance['distance_km'].round(2)
df_max_distance = df_max_distance.sort_values(by='distance_km', ascending=False)
df_max_distance.head(30).reset_index(drop=True)
```
